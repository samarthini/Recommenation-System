```python
user_books_df=books_df.pivot_table(index='userID',
                            columns='Title',
                             values='Rating').reset_index(drop=True)
```

```python
user_books_df
```

```python
user_books_df.index=books_df.userID.unique()
```

```python
user_books_df
```

```python
# input those NaNS with '0' values
user_books_df.fillna(0,inplace=True)
```

```python
user_books_df
```

```python
# calculating cosine similarity b/w users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim=1-pairwise_distances(user_books_df.values,
                             metric='cosine')
```

```python
user_sim
```

```python
# store the results in a dataframe
user_sim_df=pd.DataFrame(user_sim)
```

```python
#set the index & columns names to userids
user_sim_df.index=books_df.userID.unique()
user_sim_df.columns=books_df.userID.unique()
```

```python
user_sim_df
```

```python
np.fill_diagonal(user_sim,0)
user_sim_df
```

```python
#most similar users
user_sim_df.idxmax(axis=1)
```
